```python
# Generate a summary statistics table of mean, median, 
# Use groupby and summary statistical methods to calculate the following properties of each sale 


mean = clean_data['Sale Price'].groupby(clean_data ['Make']).mean()
median = clean_data['Sale Price'].groupby(clean_data ['Make']).median()
max = clean_data['Sale Price'].groupby(clean_data ['Make']).max()
min = clean_data['Sale Price'].groupby(clean_data['Make']).min()


```

```python
summary_sale = pd.DataFrame({
    'car sale mean': mean,
    'car sale median': median,
    'car sale max': max,
    'car sale min': min
})

summary_sale
```
